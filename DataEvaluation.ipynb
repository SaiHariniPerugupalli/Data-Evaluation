```python
#loading the dataset into boston
from sklearn import datasets
import pandas as pd
boston = datasets.load_boston()
df= pd.DataFrame(boston.data,columns=boston.feature_names)
df['target'] = pd.Series(boston.target)
df.head()
```

```python
#Housing Price Analysis
print("Summary statistics for selected features(AGE, NOX, DIS, target) from the given dataset")
dataframe=pd.DataFrame()
for i in ['AGE','NOX','DIS','target']:
    dataframe[i]=df[i]
dataframe.describe()
```

```python
#Scatter Plots
import matplotlib.lines as mlines
def scatter_plot(x,y):
    features = df
    plt.scatter(features[x], features[y], alpha=0.2,  c=boston.target, cmap='viridis', marker="*", )
    plt.title(x+'vs'+y)
    plt.ylabel('Target(Price in thosands)');
    plt.xlabel(x);
    plt.show()
```
